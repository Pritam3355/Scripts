```python
from nsetools import Nse
nse = Nse()
```

```python
tickers = ["infy","tcs","wipro","accenture","hdfc"]

quote_df = pd.DataFrame()

for ticker in tickers:
    try:
        quote = nse.get_quote(ticker)
        quote_df[quote["companyName"]]=pd.Series(quote,dtype=str)
    except:
        print(f"Stock symbol for {ticker} is : {nse.is_valid_code(ticker)}")
```

```python
quote_df.index
```

```python
Symbol = quote_df.loc['symbol']
High = quote_df.loc['dayHigh'].astype('float')
Low = quote_df.loc['dayLow'].astype('float')
Volume = quote_df.loc['totalTradedVolume'].astype('float')
Close = quote_df.loc['closePrice'].astype('float')
Open = quote_df.loc['open'].astype('float')
Code = quote_df.loc['isinCode']

df = pd.concat([Symbol,Open,Close,High,Low,Volume,Code],axis=1)
df.rename(columns={'symbol':'Symbol','open':'Open','closePrice':'Close','dayHigh':'High',
                   'dayLow':'Low','totalTradedVolume':'Volume','isinCode':'Code'},inplace=True)
df = df.reset_index(drop=True)
df
```

```python
stk_codes = nse.get_stock_codes()
symbol,company = [],[]
for k,v in stk_codes.items():
    symbol.append(k)
    company.append(v)
symbol = pd.Series(symbol)
company = pd.Series(company)
stock_codes = pd.concat([symbol,company],axis=1)
stock_codes = stock_codes.reset_index(drop=True)
stock_codes.drop(0,inplace=True)
stock_codes = stock_codes.reset_index(drop=True)
stock_codes.rename(columns={0:'Symbol',1:'Company'},inplace=True)
stock_codes
```

```python
index_codes = pd.DataFrame(pd.Series(nse.get_index_list(),name="Index"))
index_codes.head(5)
```

```python
top_gainers = pd.DataFrame(nse.get_top_gainers())
top_gainers.rename(columns={'symbol':'Symbol','openPrice':'Open','closePrice':'Close','highPrice':'High',
                   'dayLow':'Low','tradedQuantity':'Volume','ltp':'Last_Traded_Price',
                   'turnoverInLakhs':'Revenue','netPrice':'Net_Price',
                    'previousPrice':'Previous_Price'},inplace=True)
top_gainers.drop(['series','lastCorpAnnouncementDate','lastCorpAnnouncement'],axis=1,inplace=True)
top_gainers
```

```python
index_info = pd.DataFrame(nse.get_advances_declines())
last_price,change,pct_change = [],[],[]
for idx in list(index_info['indice']):
    try:
        idx_quote = nse.get_index_quote(idx)
        last_price.append(idx_quote['lastPrice'])
        change.append(idx_quote['change'])
        pct_change.append(idx_quote['pChange'])
    except:
        last_price.append(np.NaN)
        change.append(np.NaN)
        pct_change.append(np.NaN)

index_info['Last Price'] = last_price
index_info['Change'] = change
index_info['% Change'] = pct_change

index_info
```

```python
from nsepy import get_history
from nsepy.derivatives import get_expiry_date
from nsepy import get_index_pe_history
from nsepy import get_rbi_ref_history
from nsepy.history import get_price_list

from datetime import datetime
from datetime import date
```

```python
#ticker_indices = ['NIFTY 50','NIFTY BANK']
ticker_indices = ['NIFTY 50']
start_date = '1/1/2021'
end_date = '1/12/2022'
```

```python
indices_df = pd.DataFrame()
for idx in ticker_indices:
    idx_data = get_history(symbol=idx, start=pd.to_datetime(start_date),
                           end=pd.to_datetime(end_date), index = True)
    idx_data = pd.DataFrame(idx_data)
    idx_data['Index_Name'] = idx
    indices_df = pd.concat([indices_df,idx_data])
    
indices_df

# indices_df.to_csv('Nifty50_Data_2022.csv', index=True)
```
